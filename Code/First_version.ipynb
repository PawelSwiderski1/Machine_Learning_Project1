Znormalizowanie za pomocą podzielenia przez średnią, oraz sprawdzenie czy po normalizacji wychodzą jako średnie same 1

```python
X_train.dtypes.unique()
```

```python
Q1 = X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)
IQR = Q3 - Q1
print(((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).sum())
print("W sumie wszystkich outlierów jest: ", ((X_train < (Q1 - 1.5 * IQR)) | (X_train > (Q3 + 1.5 * IQR))).sum().sum())
```

Spróbujemy dobrać inne granice klasyfikacji outlierów i zobaczyć czy nadal jest ich tak dużo. Zmieńmy współczynnik przy IQR nas 2.5

```python
Q1 = X_train.quantile(0.25)
Q3 = X_train.quantile(0.75)
IQR = Q3 - Q1
print(((X_train < (Q1 - 2.5 * IQR)) | (X_train > (Q3 + 2.5 * IQR))).sum())
print("W sumie wszystkich outlierów jest: ", ((X_train < (Q1 - 2.5 * IQR)) | (X_train > (Q3 + 2.5 * IQR))).sum().sum())
```

Nadal outlierów jest bardzo dużo, dlatego usuwanie nie wchodzi w grę. Uważamy, że najlepszym sposobem będzie przeskalowanie wszystkich wartości z użyciem Robust Scaling'u, który idealnie nadaje się do przypadków, w których mamy do czynienia z dużą ilością outlierów i chcemy przeskalować wartości i nie chcemy, żeby outliery "przyćmiły" resztę punktów.

Szukanie sensownego variance thresholdu

```python
from matplotlib import pyplot as plt
```

Najpierw musimy znormalizować dane

```python
from sklearn.preprocessing import normalize
normalize = normalize(X_train)
normalize
```

```python
data_scaled = pd.DataFrame(normalize)
data_scaled.var()
```

```python
variance = data_scaled.var()
columns = X_train.columns
```

Variable będzie przechowywać kolumny których wariancja jest większa niż treshold

```python
variable = [ ]

for i in range(0,len(variance)):
    if variance[i]>0.001: #setting the threshold at 0.001
        variable.append(columns[i])
```

Sprawdzamy po każdym nowym tresholdzie ile kolumn spełniałoby takie wymaganie

```python
len(variable)
```

```python
variable = [ ]

for i in range(0,len(variance)):
    if variance[i]>0.0001: #setting the threshold at 0.0001
        variable.append(columns[i])
```

```python
len(variable)
```

```python
variable = [ ]

for i in range(0,len(variance)):
    if variance[i]>0.00001: #setting the threshold at 0.00001
        variable.append(columns[i])
```

```python
len(variable)
```

Jak widać nawet przy małym treshold nie zostaje dużo kolumn, dlatego nie wiem czy opłaca się usuwać kolumny z małą wariancją
